```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load dataset
data = pd.read_csv("users.csv")

# Show first few rows
print(data.head())

# Select features (ignore User_ID)
X = data[['Activity', 'Time_Spent', 'Posts', 'Likes']]

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=0)
data['Cluster'] = kmeans.fit_predict(X)

# Show results
print(data)

# Plot clusters (2D)
plt.scatter(data['Activity'], data['Time_Spent'], c=data['Cluster'])
plt.xlabel("Activity")
plt.ylabel("Time Spent")
plt.title("User Segmentation")
plt.show()
```
